```python
import pandas as pd
ODP = pd.read_csv('6031364e979e813c.csv')
```

```python
ODP.head()
```

```python
from matplotlib import pyplot as plt
ODP["date"] = pd.to_datetime(ODP['date'])
print(ODP[ODP["date"]=='2008-11-28'])
pl= ODP.plot(kind='line',x='date',y='Annual Return',color='black')
plt.scatter('2008-11-28', ODP[ODP["date"]=='2008-11-28']['Annual Return'], color="red")
pl.text('2008-11-28', ODP[ODP["date"]=='2008-11-28']['Annual Return'], "2008-11-28", horizontalalignment='right', size='medium', color='red', weight='semibold')
plt.title('ODP Annual Return')
plt.show()
```
