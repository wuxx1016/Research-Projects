```python
import pandas as pd
data = pd.read_csv('939e5cf42f545d8f.csv')
```

```python
data.head()
```

```python
data["date"] = pd.to_datetime(data['date'])
WMB = data[data['TICKER'] == 'WMB']
WMB.head()
```

```python
from matplotlib import pyplot as plt
print(WMB[WMB["date"]=='2002-07-31'])
pl = WMB.plot(kind='line',x='date',y='Annual Return',color='black')
plt.scatter('2002-07-31', WMB[WMB["date"]=='2002-07-31']['Annual Return'], color="red")
plt.text('2002-07-31', WMB[WMB["date"]=='2002-07-31']['Annual Return'], "2002-07-31", horizontalalignment='right', size='medium', color='red', weight='semibold')
plt.title('WMB Annual Return')
#plt.show()


WMB = WMB.set_index(WMB['date'])
WMB1 = WMB.loc['2001-01-01':'2003-12-31',:]
pl1 = WMB1.plot(kind='line',x='date',y='Annual Return',color='black')
plt.scatter('2002-07-31',WMB[WMB["date"] == '2002-07-31']['Annual Return'], color="red")
pl1.text('2002-07-31', WMB[WMB["date"]== '2002-07-31']['Annual Return'], '2002-07-31', horizontalalignment='right', size='medium', color='red', weight='semibold')
plt.title('WMB Annual Return')
plt.show()
```

```python
print(WMB[WMB["date"]=='2002-07-31'])
pl= WMB.plot(kind='line',x='date',y='PRC',color='black')
plt.scatter('2002-07-31', WMB[WMB["date"]=='2002-07-31']['PRC'], color="red")
pl.text('2002-07-31', WMB[WMB["date"]=='2002-07-31']['PRC'], "2002-07-31", horizontalalignment='right', size='medium', color='red', weight='semibold')
plt.title('WMB Price')
plt.show()
```
